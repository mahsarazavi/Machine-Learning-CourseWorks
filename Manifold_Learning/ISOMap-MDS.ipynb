```
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from matplotlib import pyplot as plt
import plotly.express as px
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.manifold import Isomap
from sklearn.manifold import MDS
```

Read data

```
data = pd.read_csv("SwissRoll.csv")

df = pd.DataFrame(data)
x = df.drop('color', 1)
print(np.shape(x))
```

Ploting data

```
fig = px.scatter_3d(x, x='x0', y='x1', z='x2',
              color=df['color'])
fig.show()
```

LLE on data

```
embedding = LocallyLinearEmbedding(n_components=2)
x_transformed = embedding.fit_transform(x)


plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.scatter(
    x_transformed[:,0],
    x_transformed[:,1],
    c=df['color']
    
)
```

Isomap on data

```
embedding = Isomap(n_components=2)
x_transformed = embedding.fit_transform(x)

plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.scatter(
    x_transformed[:,0],
    x_transformed[:,1],
    c=df['color']
    
)
```

MDS on data

```
embedding = MDS(n_components=2)
x_transformed = embedding.fit_transform(x)

plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.scatter(
    x_transformed[:,0],
    x_transformed[:,1],
    c=df['color']
    
)
```

Add noise to data

```

mu, sigma = 0, 0.5 

noise = np.random.normal(mu, sigma, [1500,3]) 

noisy = x + noise
```

Plot noisy data

```
fig = px.scatter_3d(noisy, x='x0', y='x1', z='x2',
              color=df['color'])
fig.show()
```

LLE on noisy data

```
embedding = LocallyLinearEmbedding(n_components=2)
noisy_transformed = embedding.fit_transform(noisy)


plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.scatter(
    noisy_transformed[:,0],
    noisy_transformed[:,1],
    c=df['color']
    
)
```

Isomap on noisy data

```
embedding = Isomap(n_components=2)
noisy_transformed = embedding.fit_transform(noisy)

plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.scatter(
    noisy_transformed[:,0],
    noisy_transformed[:,1],
    c=df['color']
    
)
```

MDS on noisy data

```
embedding = MDS(n_components=2)
noisy_transformed = embedding.fit_transform(noisy)

plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.scatter(
    noisy_transformed[:,0],
    noisy_transformed[:,1],
    c=df['color']
    
)
```
